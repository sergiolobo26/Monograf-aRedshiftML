```python
#funtion to select specific data from the original data_set


def select_dataset(name_to_select, target_column, true_data, tar_data):
    
    '''
    returns data_set_1 and data-set_2 with only the data corresponding to name_to_select
    '''
    n_rows = len(true_data[1][target_column][:])

    true_array_to_return = []
    tar_array_to_return = []
    for i in range(n_rows):
        name_in_file = true_data[1][target_column][i]
        if name_to_select == name_in_file:
            true_array_to_return.append(true_data[1][i])
            tar_array_to_return.append(tar_data[1][i])
            
    true_array_to_return = array(true_array_to_return)
    tar_array_to_return = array(tar_array_to_return)
    
    return true_array_to_return, tar_array_to_return

    
```

# Descripcin del algoritmo:
1. flux_class : arreglo de tamao (#de datos, 5 = # de flujos )
2. se crea bin_edges: lista de (5, # bins + 1), representa los bins en que se divide cada flujo para calcular la distribucion de los datos. 
3. Se crea class_mask: arreglo de (# flux bins = 50, # flux classes = 5, # de datos). Este arreglo contiene True o False dependiendo de la el
bin en el que se encuentra un dato. Por ejemplo: El dato 2000, tiene un Flujo_G de 1.5, por lo tanto, pertene al bin # 10 de esta clase, entonces
class_mask[:,1,2000] es un arreglo de longitud 50 con False en todas sus casilas excepto en la 10. 
4. Se llena el arreglo flux_class: se recorre el arreglo class_mask, y se encuentran las casillas de True para 'clasificar' cada dato en una 'clase'
o bin de cada uno de los Flujos. Se asigna as: flux_class[i, j] = where(class_mask[:, j , i] == True)[0]. Por lo tanto, flux_class esta lleno de enteros
de 0 a # bins -1 = 49.
5. Luego se agrega flux_class a fluxes para obtener una tabla con la clasificacion de cada dato por bin y por tipo de flujo. 
6. Se construyen histogramas de alpha por cada bin de cada clase. 
7. Por cada bin, se calcula mean y std, luego se grafica. Esto es lo que debemos aprender. 


```python
# Crear columna por flux_ y a cada fila asignarle la 'clase' a la que pertenece. Cada clase es un intervalo de dicho flux. 
flux_class = zeros((len(fluxes_array), 5))#, ('Cl_FR', 'i4'),('Cl_FZ', 'i4'),('Cl_FW1', 'i4'),('Cl_FW2', 'i4')])

histo = []
bin_edges = []

for i, name in enumerate(flux_names):
    histo_temp, bin_edges_temp = np.histogram(fluxes[name], 50,  density=False)
    bin_edges_temp[-1] = bin_edges_temp[-1]*1.01
    histo.append(histo_temp)
    bin_edges.append(bin_edges_temp)

class_mask = []
for i in range(shape(bin_edges)[1] - 1):
    name_mask = []
    for j, name in enumerate(flux_names):
            temp_1 = array(fluxes[name] >= bin_edges[j][i])
            temp_2 = array(fluxes[name] < bin_edges[j][i + 1])
            temp_3 = logical_and(temp_1, temp_2)
            name_mask.append(temp_3)
    class_mask.append(name_mask)    
class_mask = array(class_mask)   
    

```

```python
#check class_mask: it sould have only one true cell.
# class_mask = (50 , 5 , # datos)
def check_class_mask():
    for i in range(5):
        for j in range(889316):
            bin_vector = class_mask[:, i, j]  #dato j de la clase 1
            if where(bin_vector)[0].size != 1:
                print('Nmero de Trues: {}'.fomat(where(bin_vector)[0].size))
                print(i, j)
                break
    print('Todo ok')
check_class_mask()
```

# Histogramas de alpha per bin
Notese que al aumentar el bin de 0 a 8 la distribucin se hace ms ancha

```python
figure(figsize=(20,10), dpi = 200)
for i, name in enumerate(flux_names):
    subplot(2,3,1+i)
    
    legend_names = []
    for j in range(0, 10, 2):
        class_g = fluxes['class_' + name] == j
        alpha_name = abs(fluxes['alpha'][class_g])
        n, bins, patches = hist(log(alpha_name), 50, (min(log(alpha_name)), max(log(alpha_name))), histtype = 'step', log=True, align = 'mid')
        legend_names.append('bin {}'.format(j))
    legend(legend_names)
    title('LOG {}'.format(name))
    xlabel(r'Log $\alpha$')
    ylabel('Count')
```

```python
# compute the \mu and \sigma of every bean per flux class
# keep in mind: este arreglo contiene NaNs (cuando no hay datos en un bin)por lo tanto, si se opera con el, aparecern todo como NaN
log_std_mean_array = zeros((shape(bin_edges)[1] - 1, 5, 2))
for i, name in enumerate(flux_names):
    class_name = 'class_' + name
    for j in range(shape(bin_edges)[1] - 1):
        array_class = fluxes[class_name] == j
        alpha_temp = log10(fluxes['alpha'][array_class])
        
        if alpha_temp.size:
            mean = alpha_temp.mean()
            std = alpha_temp.std()
            
            log_std_mean_array[j, i, 0] = mean
            log_std_mean_array[j, i, 1] = std

        else:
            log_std_mean_array[j, i, 0] = nan
            log_std_mean_array[j, i, 1] = nan
        

        
```

```python

figure(figsize=(20,10), dpi = 200)
for i, name in enumerate(flux_names):
    subplot(2,3,1+i)
    x = linspace(min(bin_edges[i][0:-1]), max(bin_edges[i][0:-1]), 20)
    y = zeros(len(x))
    std_i = log_std_mean_array[:,i, 1]
    error_i = [std_i[j]/sqrt(histo[i][j])  for j in range(shape(bin_edges)[1] - 1)]
    plot(x, y, '--')
    errorbar(bin_edges[i][0:-1], log_std_mean_array[:,i, 0], yerr= error_i, fmt='ro', markersize=2, ecolor='g', capthick=10)
    xlabel('LOG {}'.format(name))
    ylabel(r'$\mu$ (Log $\alpha$)')

#GRAFICAS DE SIGMA
figure(figsize=(20,10))
for i, name in enumerate(flux_names):
    subplot(2,3,1+i)
    scatter(bin_edges[i][0:-1], log_std_mean_array[:,i, 1], marker='+', s=8, c='m')
    xlabel('LOG {}'.format(name))
    ylabel(r'$\sigma$ (Log $\alpha$)')

#GR+AFICA DE SIGMA/RAIZ(N)
figure(figsize=(20,10))
for i, name in enumerate(flux_names):
    subplot(2,3,1+i)
    scatter(bin_edges[i][0:-1], log_std_mean_array[:,i, 1]/sqrt(histo[i][:]) , marker='+', s=8, c='m')
    xlabel('LOG {}'.format(name))
    ylabel(r'$\sigma/\sqrt{n}$ (Log $\alpha$)')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

def transform_to_array(rec_array):
    
    dim_1 = len(rec_array)
    dim_2 = len(rec_array[0])
    temp = zeros((dim_1, dim_2))

    for i in range(dim_1):
        for j in range(dim_2):
            temp[i, j] = rec_array[i][j]
        
    dict_of_names = {name:i  for i, name in enumerate(rec_array.dtype.names)} #index of the names in the new array
    return temp, dict_of_names

def omit_columns_of_array(columns_to_omit, x_array):
    
    dim_1 = x_array.shape[0] #rows
    dim_2 = x_array.shape[1] #columns
    
    n = arange(dim_2)
    
    for i in columns_to_omit:
        msk = n != i
        n = n[msk]
    return x_array[:, n]


def obtain_results(X, y, regressor, z_col_idx, trueZ_col_idx,  scale = False):
    #z_col_idx is de index of the z column in the X matrix
    #trueZ_col_idx is de index of the TRUEZ column in the X matrix
    print('Splitting data set \n')
    X_train_i, X_test_i, y_train, y_test = train_test_split(X, y , test_size = 0.15, random_state=0)
    
    
    Z_train = X_train_i[:, [z_col_idx, trueZ_col_idx]]
    Z_test = X_test_i[:, [z_col_idx, trueZ_col_idx]]
    print(shape(Z_train), shape(Z_test))
    
    X_train = omit_columns_of_array([z_col_idx, trueZ_col_idx], X_train_i)
    X_test = omit_columns_of_array([z_col_idx, trueZ_col_idx], X_test_i)
    
    if scale:
        y_train = reshape(y_train, (len(y_train), 1))
        y_test = reshape(y_test, (len(y_test), 1))
        print('Scaling data \n')
        sc_X = StandardScaler()
        X_train = sc_X.fit_transform(X_train)
        X_test = sc_X.transform(X_test)
        sc_y = StandardScaler()
        y_train = sc_y.fit_transform(y_train)
        y_test = sc_y.transform(y_test) 
        y_train = reshape(y_train, (len(y_train)))
        y_test = reshape(y_test, (len(y_test)))
        
    print('Fitting regressor \n')
    regressor.fit(X_train, y_train)
    predictions_test = regressor.predict(X_test)
    predictions_train = regressor.predict(X_train)
    print('--- Regressor fitted --- \n')
    if scale: 
        predictions_test = sc_y.inverse_transform(predictions_test)
        predictions_train = sc_y.inverse_transform(predictions_train)
    
    print('n_Features = ', regressor.n_features_)
    print('Making predictions:')
    
    
    predicted_redshifts_test = multiply(predictions_test, Z_test[:, 0])
    predicted_redshifts_train = multiply(predictions_train, Z_train[:, 0])
    
    
    x = linspace(0, 100, 100)
    print('Plots: \n')
    
    figure(figsize=(20, 20))
    subplot(2, 2, 1)
    scatter(predicted_redshifts_test, Z_test[:, 1], marker='+', s=8, c='m')
    plot(x, x, '--', color = 'r' )
    #scatter(predictions_test, y_test)
    xlabel('Z predicts')
    ylabel('True Z')
    xlim(0.0, 3.0)
    ylim(0.0, 1)
    
    title('Test set')
    
    subplot(2, 2, 2)
    plot(x, x, '--', color = 'r' )
    scatter(predicted_redshifts_train, Z_train[:, 1],marker='+', s=8, c='m' )
    #scatter(predictions_train, y_train)
    xlabel('Z predicts')
    ylabel('True Z')
    xlim(0.0, 3.0)
    ylim(0.0, 1)
        
    title('Train set')
    
    figure(figsize=(20, 20))
    subplot(2, 2, 3)
    scatter(Z_test[:, 0], Z_test[:, 1], marker='+', s=8, c='m')
    plot(x, x, '--', color = 'r' )
    #scatter(predictions_test, y_test)
    xlabel('Z Tar')
    ylabel('True Z')
    xlim(0.0, 3.0)
    ylim(0.0, 1)
    
    title('Test set')
    
    subplot(2, 2, 4)
    plot(x, x, '--', color = 'r' )
    scatter(Z_train[:, 0], Z_train[:, 1],marker='+', s=8, c='m' )
    #scatter(predictions_train, y_train)
    xlabel('Z Tar')
    ylabel('True Z')
    xlim(0.0, 3.0)
    ylim(0.0, 1)

    title('Train set')
```

```python
#SVR
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import accuracy_score, f1_score, r2_score,  precision_score, recall_score

X_entrada, dict_of_names = transform_to_array(fluxes[['FLUX_G', 'FLUX_R', 'FLUX_Z', 'FLUX_W1', 'FLUX_W2',  'Z', 'TRUEZ' ]])
y_entrada = fluxes['alpha']

z_col_idx = dict_of_names['Z']
trueZ_col_idx = dict_of_names['TRUEZ']

X_train_i, X_test_i, y_train, y_test = train_test_split(X_entrada, y_entrada , test_size = 0.25, random_state=0)

Z_train = X_train_i[:, [z_col_idx, trueZ_col_idx]]
Z_test = X_test_i[:, [z_col_idx, trueZ_col_idx]]

X_train = omit_columns_of_array([z_col_idx, trueZ_col_idx], X_train_i)
X_test = omit_columns_of_array([z_col_idx, trueZ_col_idx], X_test_i)

#scale
#y_train = reshape(y_train, (len(y_train), 1))
#y_test = reshape(y_test, (len(y_test), 1))
print('Scaling data \n')
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
#sc_y = StandardScaler()
#y_train = sc_y.fit_transform(y_train)
#y_test = sc_y.transform(y_test) 
#y_train = reshape(y_train, (len(y_train)))
#y_test = reshape(y_test, (len(y_test)))

tuned_parameters = [{'C': [1, 10], 'kernel': ['rbf'], 'gamma': 
    [0.01, 10], 'epsilon': [0.01, 10], 'degree': [3]}]

scores = ['r2']#, 'neg_mean_squared_error']
best_parameters = {'r2':0, 'neg_mean_squared_error':0, 'explained_variance':0}
for score in scores:
    print("# Tuning hyper-parameters for %s" % score)
    print()
    clf = GridSearchCV(SVR(), tuned_parameters, cv=KFold(n_splits =2, random_state=0 ), n_jobs = -1,
                       scoring='%s' % score, verbose = 2)
    clf.fit(X_train, y_train)

    print("Best parameters set found on development set:")
    print()
    print(clf.best_params_)
    print()
    best_parameters[score] = clf.best_params_
    print("Grid scores on development set:")
    print()
    means = clf.cv_results_['mean_test_score']
    stds = clf.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, clf.cv_results_['params']):
        print("%0.3f (+/-%0.03f) for %r"
              % (mean, std * 2, params))
    print()

    print("Detailed classification report:")
    print()
    print("The model is trained on the full development set.")
    print("The scores are computed on the full evaluation set.")
    print()
    y_true, y_pred = y_test, clf.predict(X_test)
    print(r2_score(y_true, y_pred))
    print()
```
